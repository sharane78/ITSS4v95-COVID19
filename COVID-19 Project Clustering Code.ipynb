```python
# import relevant libraries
import turicreate 
from turicreate import SArray
```

```python
# clustering_data is a dataset that includes the two relevant attributes of data: confirmed cases and fatalities
clustering_data = turicreate.SFrame('~/Desktop/ITSS4v95/Preprocessed confirmed cases and fatalities train.csv')

# SArrays created for fatalities and for confirmed cases for further data analysis
fatalities_sa = SArray('~/Desktop/ITSS4v95/march 24 preprocess fatalities.csv')
confirmedcases_sa = SArray('~/Desktop/ITSS4v95/march 24 preprocess confirmed cases.csv')
```

```python
# clustering_model alogrithm
clustering_model = turicreate.dbscan.create(clustering_data, features=None, 
                                            distance=None, radius=1.0, min_core_neighbors=10, verbose=True)
```

```python
# summary of the clustering model
clustering_model.summary(output=None)
```

```python
# plot of the two SArrays - confirmed cases and fatalities
turicreate.plot(confirmedcases_sa, fatalities_sa, xlabel='ConfirmedCases', ylabel='Fatalities', title='March 24th')
```

```python
# plotting confirmedcases SArray
confirmedcases_sa.plot(title='March 24 Global Cases', xlabel='Frequency', ylabel='Confirmed cases in each country on March 24')
```

```python
#plotting fatalities SArrays
fatalities_sa.plot(title='March 24 Global Fatalities', xlabel='Frequency', ylabel='Fatalities in each country on March 24')
```
